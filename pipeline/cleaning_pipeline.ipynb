```python
import os
import pandas as pd


# Add the Graphviz bin directory to PATH
os.environ["PATH"] += os.pathsep + r"C:\Program Files\Graphviz\bin"

from graphviz import Digraph
```

```python
flowchart = Digraph()
# flowchart.attr(rankdir ='LR') #left to right flowchart

flowchart.node('A','Start')
flowchart.node('B', 'Load Data')
flowchart.node('C', 'Remove Inconsistent Recipes')
flowchart.node('D', 'Normalize Text')
flowchart.node('E', 'Whitespace Management')
flowchart.node('F', 'Validate Recipe Structure')
flowchart.node('G', 'Standardize Ingredients')
flowchart.node('H', 'Deduplicate Recipes')
flowchart.node('I', 'Check Spelling')
flowchart.node('J', 'Save Clean Data')
flowchart.node('K', 'End')

flowchart.edges(['AB','BC','BC','CD','DE','EF','FG','GH','HI','IJ','JK'])

# flowchart.edges(['AB', 'BC', 'CD', 'DE', 'EF', 'FG', 'GH', 'HI', 'IJ'])
# flowchart.edge('J', 'K', 'No Errors')  
# flowchart.edge('K', 'L')  
# flowchart.edge('I', 'J', 'Errors Found') 

flowchart.render('cleaning_pipeline', format='png')
```

```python
docs = 'C:\\Users\\oyku_\\Desktop\\Recipe Collector\\Recipe-Collector\\pipeline\\data'

dataframes = []

for filename in os.listdir(docs):
    if filename.endswith('csv'):
        file_path = os.path.join(docs,filename)
        df = pd.read_csv(file_path, on_bad_lines='skip')
        dataframes.append((filename, df))
print(f"Number of DataFrames loaded: {len(dataframes)}")
```
