```python
import requests
from bs4 import BeautifulSoup

def get_collection_links(base_url):

    collection_links = []
    try:
        response = requests.get(base_url)
        response.raise_for_status() 
        soup = BeautifulSoup(response.text, 'html.parser')

        for link in soup.find_all('a', href=True):
            if '/foods-and-drinks/' in link['href']: 
                collection_url = f"{link['href']}"
                if collection_url not in collection_links:  # Avoid duplicates NOT WORKING
                    collection_links.append(collection_url)

        print(f"Found {len(collection_links)} collections.")
    except requests.exceptions.RequestException as e:
        print(f"Error fetching collection page: {e}")
    
    return collection_links

def get_recipe_links_from_collection(collection_url):

    pdf_links = []
    try:
        response = requests.get(collection_url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')

        for link in soup.find_all('a', href=True, class_='Libros_Boton_Dos'):
            if  link['href']:
                pdf_url = link['href']
                if pdf_url not in pdf_links:
                    pdf_links.append(pdf_url)

        print(f" {collection_url}")
    except requests.exceptions.RequestException as e:
        print(f"Error fetching recipes from {collection_url}: {e}")
    
    return pdf_links

def scrape_morrisons_recipes(base_url):

    all_pdf_links = []

    collection_links = get_collection_links(base_url)

    for collection_url in collection_links:
        pdf_links = get_recipe_links_from_collection(collection_url)
        all_pdf_links.extend(pdf_links)  # Add to the overall list

    return all_pdf_links

if __name__ == "__main__":
    base_url = 'https://www.infobooks.org/free-pdf-books/foods-and-drinks/'
    all_pdf_links = scrape_morrisons_recipes(base_url)

    for url in all_pdf_links:
        print(url)

```

```python
import spacy
import pandas as pd
import pdfplumber
import re
import os

language_processor = spacy.load("en_core_web_sm")

folder_path = "C:\\Users\\oyku_\\Desktop\\Recipe Collector\\Recipe-Collector\\pdf_collectors\\pdfs_from_infobooks"

cols = ["Title", "Ingeredients", "Instructions"]
rows = []

def recipe_info_title(text):

    lines = text.split("\n")
    for line in lines[:3]:
        if line.isupper() or re.match(r'^[A-Z][a-z]+', line):
            return line.strip()

    return lines[0].strip()

def recipe_info_ing(text):

    ing_keywords = ['ingredients','what you need', 'you will need', 'needs', 'stuff you need']
    ins_keywords = ['instructions', 'method', 'directions', 'how to', 'how', 'preperations' ]

    ingredients = []
    lines = text.split("\n")
    collect_ing = False

    for line in lines:

        if any(keyword in line.lower()for keyword in ing_keywords):
            collect_ing = True 

            continue


        if any(keyword in line.lower()for keyword in ins_keywords):
            break
        
        if collect_ing:
            ingredients.append(line.strip())
    return "\n".join([ing for ing in ingredients if ing.strip()])

def recipe_info_ins(text):

    ins_keywords = ['instructions', 'method', 'directions', 'how to', 'how', 'preperations' ]
    instructions =[]
    lines = text.split("\n")

    collect_ins = False

    for line in lines:

        if any(keyword in line.lower() for keyword in ins_keywords):
            
            collect_ins = True 
            continue

        if collect_ins:
            instructions.append(line.strip())
    return "\n".join([ing for ing in instructions if ing.strip()])


for file_name in os.listdir(folder_path):
    
    if file_name.endswith('.pdf'):
        pdf_path = os.path.join(folder_path, file_name) #itterate all pdfs

        try:
            with pdfplumber.open(pdf_path) as pdf:

                pdf_text = ""
                
                for page in pdf.pages:
                    pdf_text += page.extract_text()

                document = language_processor(pdf_text)

                title = recipe_info_title(pdf_text)
                ingredients = recipe_info_ing(pdf_text)
                instructions = recipe_info_ins(pdf_text)

                rows.append([title, ingredients, instructions])
        except Exception as e:

            print(f"Failed to process {file_name}: {e}")
            continue

df = pd.DataFrame(rows, columns = cols)

print(df)
```
