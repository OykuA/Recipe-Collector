```python
import PyPDF2
import pandas as pd
import requests
from io import BytesIO
import re

def parse_recipes(ingredient_text, instruction_text):
    recipes = []

    ingredient_lines = ingredient_text.strip().split('\n')
    instruction_lines = instruction_text.strip().split('\n')

    title = find_title(instruction_lines)

    if ingredient_lines and instruction_lines:
        ingredients = ' '.join(ingredient_lines)  # Join all ingredients
        instructions = ' '.join(instruction_lines[1:])  # Join instructions after the title
        recipes.append({'Title': title, 'Ingredients': ingredients, 'Instructions': instructions})

    return recipes

def find_title(instruction_lines):
    for line in instruction_lines:
        stripped_line = line.strip()
        if stripped_line.isupper() and is_valid_title(stripped_line):  # Check for all uppercase and valid title
            return stripped_line  
    
    return 'No Title Found'

def is_valid_title(line):
    excluded_keywords = [
        "IN THE KIT YOU WILL NEED", "STORAGE", "VIDEO", "SKILL LEVEL", 
        "SHARE YOUR BAKEDIN EXPERIENCE WITH US", "@BAKEDINCAKES", 
        "PREP TIME", "BAKE TIME", "MAKES", "SERVINGS", 
        "PREP CHILL TIME", "MY TOP TIP", "BAKE TIME MINS"
    ]
    
    # Normalize line to lowercase for comparison
    normalized_line = re.sub(r'\s+', ' ', line).strip().lower()
    
    for keyword in excluded_keywords:
        if keyword.lower() in normalized_line:
            return False

    return True  

def download_pdf(url):
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=headers)
    response.raise_for_status()  # Check for request errors
    return BytesIO(response.content)  # Return as a BytesIO object 

def extract_recipes(pdf_url):
    pdf_file = download_pdf(pdf_url)

    pdf_reader = PyPDF2.PdfReader(pdf_file)

    all_recipes = []

    total_pages = len(pdf_reader.pages)

    if total_pages < 2:
        print(f"PDF {pdf_url} does not have enough pages.")
        return []

    for page_num in range(0, total_pages, 2):  # Process pairs of pages
        if page_num < total_pages:
            ingredient_page = pdf_reader.pages[page_num]
            ingredient_text = ingredient_page.extract_text() or ''
        
        if page_num + 1 < total_pages:
            instruction_page = pdf_reader.pages[page_num + 1]
            instruction_text = instruction_page.extract_text() or ''
        else:
            instruction_text = ''
        
        recipes = parse_recipes(ingredient_text, instruction_text)
        all_recipes.extend(recipes)

    return all_recipes

pdf_urls = [
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Almond_Apricot_Biscotti.pdf?v=1695300808",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Banoffee_bars.pdf?v=1695300807",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Banoffee_Cookies.pdf?v=1695300808",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Belgian_Buns.pdf?v=1695300807",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/blueberry_jammy_squares.pdf?v=1695300808",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Butterscotch_Brownie.pdf?v=1695300808",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Butterscotch_Brownie.pdf?v=1695300808",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Caramel_Latte_Cake.pdf?v=1695300808",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Chocolate_Vanilla_Marble_Loaf.pdf?v=1695300808",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Chocolate-Brookies.pdf?v=1695300808",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Custard_cream_biscuits.pdf?v=1695300808",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/jaffa_loaf.pdf?v=1695300808",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Lemon_Curd_Muffins.pdf?v=1695300807",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Lemon-Drizzle-Loaf.pdf?v=1695300808",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/lime_and_coconut_bars.pdf?v=1695300807",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Malted_Millionaire_Squares.pdf?v=1695300807",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/millionnaires_shortbread.pdf?v=1695300808",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Mini-Victoria-Sponges.pdf?v=1695300809",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Strawberry_Almond_Squares.pdf?v=1695300808",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Tea_Loaf.pdf?v=1695303777",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/multigrain_rolls.pdf?v=1695716711",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Chocolate_Cheesecake_Brownies.pdf?v=1695716710",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Banana_Cashew_Loaf.pdf?v=1695716710",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Sticky_Toffee_Drip_Cake.pdf?v=1695716709",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/chocolate_and_raspberry_maracrons.pdf?v=1695716708",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/blackberry_crumble_squares.pdf?v=1695716708",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/caramel_cookie_squares.pdf?v=1695716711",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/GARLIC___rosemary_focaccia.pdf?v=1695716711",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Cinnamon_Bun_Muffins.pdf?v=1695716711",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Cherry_Bakewell_Slices.pdf?v=1695716712",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Chocolate_and_Caramel_Eclairs.pdf?v=1695716712",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Cheese_and_Garlic_Tear_and_Share.pdf?v=1695716711",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Pan_de_Agua.pdf?v=1695716712",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/date_and_walnut_bloomer.pdf?v=1695716711",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Lemon_Poppy_Seed_Muffins.pdf?v=1696330897",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Chocolate_Honeycomb_Cake.pdf?v=1696330897",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Pizza_Pinwheels.pdf?v=1696330896",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Raspberry_Eton_Mess_Cake.pdf?v=1696330896",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Tea_Loaf_d725a94f-8ab0-483e-b1cc-a639dbf8f03a.pdf?v=1696330896",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Sticky_toffee_ginger_loaf.pdf?v=1696330896",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/CARROT_CAKE_WHOOPIE_PIES.pdf?v=1696330895",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/sticky_toffee_cookie.pdf?v=1696330895",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Carrot_Cake_Cookies.pdf?v=1696330895",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Sour_cream_chocolate_cake.pdf?v=1696330895",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/toffee_apple_drip_cakea8d.pdf?v=1696330895",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Blueberry_and_White_Chocolate_Cupcakes.pdf?v=1696330894",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Dulce_de_Leche_biscuits.pdf?v=1696330893",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Sticky_toffee_fudge_cake.pdf?v=1696330893",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Double_chocolate_loaf.pdf?v=1696928935",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Greek_Yoghurt_Chocolate_Cake.pdf?v=1696928935",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Sticky-Toffee-Ginger-Loaf.pdf?v=1696928934",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Blackberry-Crumble-Coconut-Squares.pdf?v=1696928934",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/carrot_cake_biscotti.pdf?v=1696928934",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/CARROT___PECAN_SQUARES.pdf?v=1696928934",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/HONEY_WHEAT_BAGUETTES.pdf?v=1696928934",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/cheddar_and_chive_swirls.pdf?v=1696928934",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Camembert_and_cranberry_focaccia.pdf?v=1696928934",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Toffee_Apple_Muffins.pdf?v=1696928933",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Pesto_and_sun-dried_tomato_wool_bread.pdf?v=1696928931",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Pumpkin_seed_knots.pdf?v=1696928930",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Hot_Chocolate_Muffins.pdf?v=1696928930",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Sunflower_Seed_Boule.pdf?v=1697529208",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Rocky_Road_Brownie.pdf?v=1697529207",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Chocolate_Orange_Hot_Cross_Buns.pdf?v=1697529206",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Mocha_Swiss_roll.pdf?v=1697529206",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/White_Chocolate_Cranberry_Squares.pdf?v=1697529206",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/olive_and_thyme_loaf.pdf?v=1697529205",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/chocolate_orange_cookies.pdf?v=1697529205",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Chocolate-orange-marble-cake.pdf?v=1697529205",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/BLUEBERRY___CASHEW_SQUARES.pdf?v=1697529205",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/CHOCOLATE___PECAN_MILLIONAIRE_SQUARES.pdf?v=1697529205",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Orange_Almond_Cake.pdf?v=1697529205",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/junkyard_blondies.pdf?v=1697529204",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Fruity_Chocolate_Flapjacks.pdf?v=1697529203",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Khobz.pdf?v=1697529203",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Caramelised-Onion-Fougasse.pdf?v=1697529203",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Cinnamon_Pecan_Loaf.pdf?v=1698137532",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Vanilla_Chai_Cake.pdf?v=1698137530",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/swiss_farmers_loaf.pdf?v=1698137530",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Banana_caramel_muffins.pdf?v=1698137530",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/CHOCOLATE_DRIP_CAKE.pdf?v=1698137530",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Lemon-Blueberry-Biscotti.pdf?v=1698137530",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/BAKEWELL_TART.pdf?v=1698137530",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/ciabatta.pdf?v=1698137530",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/PEAR___FRANGIPANE_TART.pdf?v=1698137529",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/pecan_pie_slices.pdf?v=1698137528",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/seeded_wholemeal_cottage_loaf.pdf?v=1698137528",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/neopolitan_cupcakes.pdf?v=1698137528",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Sage_and_Onion_Loaf.pdf?v=1698137528",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/raspberry_and_white_chocolate_brownies.pdf?v=1698747859",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Raspberry___Chocolate_cake.pdf?v=1698747860",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Apple-Crumble-Custard-Loaf.pdf?v=1698747861",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/SOFT_pretzels.pdf?v=1698747860",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/RASPBERRY_RIPPLE_CAKE.pdf?v=1698747861",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Spiced_Blackberry_Muffins.pdf?v=1698747861",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/salted_caramel_loaf.pdf?v=1698747861",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/soft_breadsticks.pdf?v=1698747861",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Coconut_and_Lime_Loaf.pdf?v=1698747862",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/strawberry_shortbread_cake.pdf?v=1698747862",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Walnut_Spelt_Loaf.pdf?v=1698747861",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Mince_Pie.pdf?v=1699359450",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/APPLE___PECAN_BUNS.pdf?v=1699359451",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Gingerbread_House.pdf?v=1699359450",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/black_forest_gateau.pdf?v=1699359450",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Gingerbread_loaf.pdf?v=1699359449",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Honey-Wheat-Baguettes.pdf?v=1699359449",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Chocolate-Pannetone.pdf?v=1699359449",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Winter-Spice-Cake.pdf?v=1699359449",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Rosemary_and_Garlic_Bloomer.pdf?v=1699359447https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Rosemary_and_Garlic_Bloomer.pdf?v=1699359447",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/apple_crumble_bars.pdf?v=1699359448",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/chocolate_crinkle_cookies.pdf?v=1699359448",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Pumpkin-Sunflower-Seed-Bloomer.pdf?v=1699359448",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/White_chocolate_yule_log.pdf?v=1699949977",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Chocolate-Hazlenut-Wreath.pdf?v=1699949978",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Lemon_Viennese_Biscuits.pdf?v=1699949977",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/potato_and_rosemary_bread.pdf?v=1699949977",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Gingerbread_Latte_Cake.pdf?v=1699949978",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/chocolate_brownie_muffins.pdf?v=1699949978",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Rhubard-and-Custard-Squares.pdf?v=1699949979",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Challah.pdf?v=1699949977",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/chocolate_and_pecan_garland.pdf?v=1699949977",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/pretzel_buns.pdf?v=1699949976",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/chocolate_rudulph_cake.pdf?v=1699949978",
"https://cdn.shopify.com/s/files/1/0476/1681/9367/files/Blueberry_Muffin.pdf?v=1699949977"

]

treat_kitchen_csv = 'Treat_Kitchen.csv'

all_recipes_df = pd.DataFrame()

for pdf_url in pdf_urls:
    recipes = extract_recipes(pdf_url)
    if recipes:
        df = pd.DataFrame(recipes)
        all_recipes_df = pd.concat([all_recipes_df, df], ignore_index=True)
    else:
        print(f"No recipes found in {pdf_url}")

if not all_recipes_df.empty:
    all_recipes_df.to_csv(treat_kitchen_csv, index=False)
    print(f"Extracted {len(all_recipes_df)} recipes to {treat_kitchen_csv}")
else:
    print("No recipes were extracted from any of the PDFs.")
```
